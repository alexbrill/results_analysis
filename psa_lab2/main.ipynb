```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("data/1time.csv", dtype=float)
df1.head()
```

```python
df2 = pd.read_csv("data/2time.csv", dtype=float)
df2.head()
```

```python
df4 = pd.read_csv("data/4time.csv", dtype=float)
df4.head()
```

```python
N = 5
```

```python
df1_average = pd.DataFrame([np.mean(df1['t'][i:N+i]) for i in range(df1.shape[0] // N)],
                  index=np.unique(df1['d']), columns=["p1"])
df1_average.head()
```

```python
df2_average = pd.DataFrame([np.mean(df2['t'][i:N+i]) for i in range(df2.shape[0] // N)],
                  index=np.unique(df2['d']), columns=["p2"])
df2_average.head()
```

```python
df4_average = pd.DataFrame([np.mean(df4['t'][i:N+i]) for i in range(df4.shape[0] // N)],
                  index=np.unique(df4['d']), columns=["p4"])
df4_average.head()
```

```python
df = pd.concat([df1_average, df2_average, df4_average], axis=1)
df.head()
```

```python
idx2 = (~df[df.columns[1]].isna())
idx4 = (~df[df.columns[2]].isna())
```

```python
B = 10

fig, ax = plt.subplots()
ax.set_title(" 2x2 - 10x10")
ax.set_ylabel("Time")
ax.set_xlabel("Matrix Dimension")
ax.grid()

ax.plot(df["p1"][:B].index, df["p1"][:B], '-r')
ax.plot(df["p2"][idx2][:B].index, df["p2"][idx2][:B], '-b')
ax.plot(df["p4"][idx4][:B].index, df["p4"][idx4][:B], '-g')

ax.legend(["1 process", "2 process", "4 process"])

plt.show()
```
