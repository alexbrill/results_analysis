```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("data/square_root.csv", dtype=float)
df.head()
```

```python
fig, ax = plt.subplots()
ax.set_title("10x10 - 1000x1000")
ax.set_ylabel("Time, s")
ax.set_xlabel("Matrix Dimension, N")
ax.grid()

for columns in df.columns[1:]:
    ax.plot(df["d"], df[columns], '-')

ax.legend(["1 thread", "2 threads", "4 threads", "8 threads"])

plt.show()
```

```python
A = 0
B = 10

fig, ax = plt.subplots()
ax.set_title(f"10x10 - {B*10}x{B*10}")
ax.set_ylabel("Time, s")
ax.set_xlabel("Matrix Dimension, N")
ax.grid()

for columns in df.columns[1:]:
    ax.plot(df["d"][A:B], df[columns][A:B], '-')

ax.legend(["1 thread", "2 threads", "4 threads", "8 threads"])

plt.show()
```
